```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("Coffee Shop Sales.csv")

# Show the first few rows
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['product_category'] = df['product_category'].str.strip().str.lower()
df['product_type'] = df['product_type'].str.strip().str.lower()
df['product_detail'] = df['product_detail'].str.strip()
df['store_location'] = df['store_location'].str.strip().str.title()
df.head()
```

```python
df['transaction_time'] = df['transaction_time'].str.strip()
```

```python
df['transaction_time'] = pd.to_datetime(df['transaction_time'], format='%H:%M:%S', errors='coerce').dt.time
```

```python
df[df['transaction_time'].isnull()]
```

```python
df['transaction_qty'] = df['transaction_qty'].astype(int)
df['unit_price'] = df['unit_price'].astype(float)
```

```python
df.to_csv("coffee_shop_sales_cleaned.csv", index=False)
```
